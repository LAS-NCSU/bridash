```python
import pandas as pd 
import geopandas as geo
import fiona
```

```python
df = pd.read_csv(r'/Users/natalie_kraft/Documents/LAS/Pew2017_ITrain.csv')
shape = geo.read_file('/Users/natalie_kraft/Documents/LAS/countries.geojson')
df
```

```python
for i in range(0, len(df)):
    
    if df['Country'][i] == 'United States':
        df['Country'][i] = 'United States of America'
    
    if df['Country'][i] == 'Tanzania':
        df['Country'][i] = "United Republic of Tanzania"
```

```python
df.groupby('Country')['trust_gov'].sum() / df.groupby('Country')['trust_gov'].count()
```

```python
df.head()
```

```python
shp = pd.DataFrame()
for i, col in enumerate(df.columns):
    if type(df[col][0]) != str:
        shp[col] = df.groupby('Country')[col].sum() / df.groupby('Country')[col].count()
```

```python
shp['fav_US']
```

```python
shp = shp.drop(columns=['Unnamed: 0', 'ID'])
shp['year'] = "2017"
```

```python
shp
```

```python
shp.to_csv("/Users/natalie_kraft/Documents/LAS/AggPewShp.csv")
```

```python
export = pd.merge(shape, shp, left_on="ADMIN", right_on="Country", how="outer")
```

```python
export[['ADMIN', 'geometry', 'fav_US', 'fav_Russia', 'fav_China']].to_file("/Users/natalie_kraft/Documents/LAS/AggPew.geojson", driver = "GeoJSON")
```
