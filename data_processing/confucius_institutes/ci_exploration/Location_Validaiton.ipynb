```python
import pandas as pd
df = pd.read_csv('Confucious Institutes.csv')
df.head()
```

```python
import urllib
import bs4


open_22 = ['Open'] * len(df['Link'])
keyword_match = [None] * len(df['Link'])
keywords = ['does not exist', '404', 'closed', 'closing', 'close', 'shut down', 'transferred']

###############################################################################################

i = 0
for link in df['Link']:
    try:
        a_website = urllib.request.urlopen(link)
        a_soup = bs4.BeautifulSoup(a_website)
        website_text = a_soup.findAll(text = True)
        
        for phrase in website_text:
            for key in keywords:
                if key in phrase:
                    open_22[i] = 'Closed'
                    keyword_match[i] = key
                    break

    except:
        open_22[i] = 'Closed'

    finally:
        i += 1
        
df['Open 4/6/22'] = open_22
df['Closing Website Flag'] =  keyword_match
```

#### Determine which links are active - unreachable websites are assumed to be inactive sites

```python
import urllib.request

link_status = []
for url in df['Link']:
    print(url)
    
    if url is None or url == "" or url == 'nan':
        link_status.append('Unknown')
        
    try:
        status_code = urllib.request.urlopen(url).getcode()
        link_status.append(status_code == 200)
        
    except:
       link_status.append('False') 
```

```python
df.insert(3, 'Active Link', link_status, False)
```

```python
#print(df['Active Link'].value_counts())
links = df[df['Active Link'] == True]['Link'].to_list()
for l in links:
    print(l)
```
