## BRI Partner Integration 

In order to evaluate a difference in soft influence and economic integration between BRI partner countries and non-partner countries, we are gathering information on the countries which have signed formal MoUs with China. 

Data has been gathered from [The Green Finance and Development Center](https://greenfdc.org/countries-of-the-belt-and-road-initiative-bri). 

Information was placed in a spreadsheet explicitly for BRI countries, this script is now to merge over these results into the country data frame. 

```python
import pandas as pd 
```

```python
countries = pd.read_csv("../../data_final/countries.csv")
partners = pd.read_csv("bri_countries.csv")
```

```python
countries = countries.merge(partners, on = ['name', 'country_id'])
```

```python
# export this data back to the data final file
countries.to_csv("../../data_final/countries.csv", index=False)
```
